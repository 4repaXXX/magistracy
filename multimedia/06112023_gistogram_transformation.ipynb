### Intensity pictures

```python
import cv2
from matplotlib import pyplot as plt

def picture_gray(image, name):
    hist = cv2.calcHist([image], [0], None, [256], [0, 256])

    plt.figure()
    plt.title("Гистограмма интенсивностей " + name)
    plt.xlabel("Интенсивность")
    plt.ylabel("Количество пикселей")
    plt.plot(hist)
    plt.xlim([0, 256])
    plt.show()
    
def picture_color(image, name):
    color = ('b', 'g', 'r')
    plt.figure()
    plt.title("Гистограмма интенсивностей " + name)
    plt.xlabel("Интенсивность")
    plt.ylabel("Количество пикселей")
    for i, col in enumerate(color):
        hist = cv2.calcHist([image], [i], None, [256], [0, 256])
        plt.plot(hist, color=col)
        plt.xlim([0, 256])

    plt.show()
```

```python
import cv2
from matplotlib import pyplot as plt

image = cv2.imread('50-Shades-of-Grey.jpg')
picture_color(image, '50-Shades-of-Grey.jpg')

image = cv2.imread('th.jpeg', cv2.IMREAD_GRAYSCALE)
picture_gray(image, 'th.jpeg')
```

### Linear transformation

```python
import cv2
import gistogram_transformation as gt

lin_pic = gt.linear('50-Shades-of-Grey.jpg')

picture_color(lin_pic, '50-adaptive.jpeg')
```
