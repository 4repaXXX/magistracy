```python
import cv2
import numpy as np
import warnings
from lib.hft import table_scale
from lib.hft import gradient_correction

warnings.filterwarnings('ignore', 'QObject::moveToThread: Current thread .*')

image = cv2.imread('images/moon.jpg', cv2.IMREAD_GRAYSCALE)

laplacian = cv2.Laplacian(image, cv2.CV_64F)
# laplacian = np.abs(laplacian)
# laplacian = np.uint8(laplacian / laplacian.max() * 255)
# laplacian = table_scale(laplacian)
laplacian = gradient_correction(laplacian)

sharp_image = cv2.add(image, -(2 * laplacian))

cv2.imshow('Original Image', image)
cv2.imshow('Laplacian', laplacian)
cv2.imshow('Sharpened Image', sharp_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
from scipy.signal import convolve2d
from lib.hft import convert_to_grayscale
from lib.hft import custom_convolve2d
from lib.hft import table_scale
from lib.hft import gradient_correction
from lib.hft import laplacian_mask_operation

gray = cv2.imread('images/moon.jpg', cv2.IMREAD_GRAYSCALE)
# gray = convert_to_grayscale(image)

laplacian_kernel = laplacian_mask_operation(1)

# Применение свертки ядра Лапласиана к изображению
# laplacian = custom_convolve2d(gray, laplacian_kernel)
# laplacian = custom_convolve2d(gray, laplacian_kernel)
laplacian = convolve2d(gray, laplacian_kernel, mode='same', boundary='fill', fillvalue=0)
# laplacian = np.clip(laplacian, 0, 255).astype(np.uint8)
# laplacian = table_scale(laplacian)
laplacian = gradient_correction(laplacian)

# Повышение резкости изображения
sharp_image = gray - laplacian

# Шаг 5: Отображение результатов
cv2.imshow('Original Image', gray)
cv2.imshow('Laplacian', laplacian)
cv2.imshow('Sharpened Image', sharp_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

#### Размытый фон(цветок)

```python
import cv2
import numpy as np
from scipy.signal import convolve2d
from lib.hft import convert_to_grayscale

image = cv2.imread('images/large-flowers.jpg')
gray = convert_to_grayscale(image)

laplacian_kernel = np.array([[0, 1, 0], 
                             [1, -4, 1], 
                             [0, 1, 0]])

# Применение свертки ядра Лапласиана к изображению
laplacian = convolve2d(gray, laplacian_kernel, mode='same', boundary='fill', fillvalue=0)
laplacian = np.clip(laplacian, 0, 255).astype(np.uint8)

# Повышение резкости изображения
sharp_image = np.clip(gray - laplacian, 0, 255).astype(np.uint8)

# Шаг 5: Отображение результатов
cv2.imshow('Original Image', gray)
cv2.imshow('Laplacian', laplacian)
cv2.imshow('Sharpened Image', sharp_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
